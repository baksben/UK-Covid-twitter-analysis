#### Plot the timeline evolution for the total number of retweets

```python
import matplotlib.pyplot as plt
```

```python
# simplify date variable 

df["created_at_rec"]=pd.to_datetime(df["created_at"]).dt.date

# group dataframe

df_grouped = (
    df[['created_at_rec', 'public_metrics.retweet_count']].groupby(['created_at_rec'])
    .agg(['sum'])
)

df_grouped = df_grouped.droplevel(axis=1, level=0).reset_index()

df_grouped.head()
```

```python
# Plotting

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(df_grouped['created_at_rec'], df_grouped['sum'], color="darkgreen") # main graphing function 

ax.set_title('Timeline evolution for the total number of retweets', fontweight="bold", fontsize=14) # setting title

plt.xlabel('Date', fontweight="bold") # setting x axis
plt.ylabel('# Retweets', fontweight="bold") # setting y axis

```

#### Plot the evolution of tweets that contain any word related to the pandemic

```python
# Creating list

contains_pandemic_word=[]
for i in df['text']:
    if re.search(r'pandemic|covid-19|face mask|sars-cov-2|coronavirus|lockdown', str(i).lower()):
        contains_pandemic_word.append(1)
    else:
        contains_pandemic_word.append(0)
df['contains_pandemic_word']=contains_pandemic_word
df['contains_pandemic_word'].value_counts()
```

```python
# group dataframe

df_grouped = (
    df[['created_at_rec', 'contains_pandemic_word']].groupby(['created_at_rec'])
    .agg(['sum'])
)

df_grouped = df_grouped.droplevel(axis=1, level=0).reset_index()

df_grouped.head()
```

```python
# Plotting

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(df_grouped['created_at_rec'], df_grouped['sum'], color="darkgreen") # main graphing function 

ax.set_title('Timeline evolution for the total of words related to the pandemic', fontweight="bold", fontsize=14) # setting title

plt.xlabel('Date', fontweight="bold") # setting x axis
plt.ylabel('# Tweets', fontweight="bold") # setting y axis

```
